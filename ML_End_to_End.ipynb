```python
def confusion_matrix_manual(y_true, y_pred):
    TP = TN = FP = FN = 0

    for actual, predicted in zip(y_true, y_pred):
        if actual == 1 and predicted == 1:
            TP += 1
        elif actual == 0 and predicted == 0:
            TN += 1
        elif actual == 0 and predicted == 1:
            FP += 1
        elif actual == 1 and predicted == 0:
            FN += 1

    return TP, TN, FP, FN
```

```python
TP, TN, FP, FN = confusion_matrix_manual(y_test, rf_pred)

print("TP:", TP)
print("TN:", TN)
print("FP:", FP)
print("FN:", FN)
```

```python
accuracy = (TP + TN) / (TP + TN + FP + FN)
accuracy
```

```python
precision = TP / (TP + FP)
precision
```

```python
recall = TP / (TP + FN)
recall
```
